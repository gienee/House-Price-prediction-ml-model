```python
## Divide  the dataset into independent and dependent      ###all the columns are the independent features and only the last columns is the dependent feature only...
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split        ###split the train and test datasets,,,33% are the test data and 67%are the train datasets..and randomly take it some rows are train datasets and some rows are test datasets randomly...

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=36)
```

```python
X.shape
```

```python
X_train.shape,X_test.shape         ###total number of train and test datasets of independent features.
```

```python
from sklearn.preprocessing import StandardScaler          ###import the standardscaler to standardize the datasets. we standardize only the independent features of train and test datasets. 
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.fit_transform(X_train)        ###standardization means mean=0 and standard deviaion=1 so use .fit and then .transform is used to find the z-score then standardize values whose mean=0 and S.D=1
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)   ###in test datasets we will only use .transform because mean and s.d is already done and it will be same for all...so we jst need to use .transform o find the z_score
```

```python
from sklearn.linear_model import LinearRegression   ###import the Linear regression algorithm.
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)   ###Train the model using x_train and y_train
```

```python
regression.coef_      ###Find the regression coefficient value of each column 
```

```python
regression.intercept_              ###Find the intercept value for the dataset.
```

```python
## prediction
y_pred=regression.predict(X_test)   ### Find the predicted value by using the train dataset i.e, x_train on the  linear model and find y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error   ### In regression problem,to find the score and accuracy import mean absolute error and mean squared error.
```

```python
import numpy as np                        ###Find the mean squared error,mean absolute error and root mean squared error for the whole datasets using actual value y_test and predicted value y_pred... 
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
## Accuracy r2 and adjusted r square   ###import r2 score to find out the accuracy of the model or the performance of the model
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)           ###To find out the performance of the model use r2 score by using of actual value y_train and predicted value i.e,y_pred        
```

```python
score
```

```python
#display adjusted R-squared                      ###Also find the adusted r2 squared by using the formula bt..the difference in score between r2 square and adjusted r2 squared is very little bit...i,e, performance value doesn't differs much.
1 - (1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge       ###Here,use the Ridge algorithm and take the hyperparameter or penality as alphan=20.0,,,and  by using train datasets trained the model  
ridge=Ridge(alpha=20.0)
ridge.fit(X_train,y_train)
```

```python
y_pred=ridge.predict(X_test)    ###Find the predicted value y_pred by giving the x_test as input
```

```python
mse=mean_squared_error(y_test,y_pred)
print(mse)                                      ###Find the error by using actual value and predicted value using the concept of mse,,mae  and rmse and it automatically gives us the accuracy of the model
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso   ###Import Lasso algo for the datasets.
```

```python
lasso=Lasso(alpha=20.0)
lasso.fit(X_train,y_train)          ###Use the hyperparameter as alpha=20.0 and train the model use x_train and y_train,,(.fit) is used to train the model.
```

```python
y_pred=lasso.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)                                      ###Find the error by using actual value and predicted value using the concept of mse,mae,rmse and it automatically gives us the accuracy of the model
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet                ###Import Elasticnet algo for the datasets.
```

```python
elasticnet=ElasticNet(alpha=20.0)
elasticnet.fit(X_train,y_train)
```

```python
y_pred=elasticnet.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)             ###Find the error by using actual value and predicted value using the concept of mse,mae,rmse and it automatically gives us the accuracy of the model
print(mae)
print(np.sqrt(mse))
```
